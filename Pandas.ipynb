```python
import pandas as pd

a = pd.DataFrame({'name':['a','b','c','d','e','f'],
     'marks':[20,45,57,78,11,39]})

def pass_fail(x):
  if x>=40:
    return 'Pass'
  else:
    return 'Fail'

a['pass_fail'] = a['marks'].apply(pass_fail)  

a
```

```python
#select * from a left join b on a.column = b.column when column names are different

import pandas as pd

a = pd.read_csv(('C:/Users/SaiSantoshNarukurthi/Downloads/Cleaned Banking.csv'))

b = a[a['job']=='unemployed']

c = pd.merge(a,b, left_on = 'age', right_on = 'age') #left_on = 'age', right_on = 'age'

c[c['age'] < 30 ]
```

Concept of comparison filtering on 'Name' column:

In pandas, comparison filtering is a way to filter rows based on a comparison between a column and a scalar value. For example, if we want to filter all rows where the value in a column is greater than 5, we can use the following syntax: df[df['column'] > 5]. Similarly, if we want to filter all rows where the value in a column is equal to a certain value, we can use the following syntax: df[df['column'] == value]. In the case of string columns like 'Name', comparison filtering works based on the alphabetical order of the strings.

# Pivot table

The pivot_table() function takes three arguments:

index: This specifies the column(s) to use as the index of the pivot table. In this case, we want to use the 'Gender' column as the index.
values: This specifies the column(s) to use as the values in the pivot table. In this case, we want to use the 'Salary' column as the values.
aggfunc: This specifies the aggregation function to use when multiple values are associated with a single index. In this case, we want to use the 'mean' function to calculate the average salary.

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 'Fred'],
        'Gender': ['F', 'M', 'M', 'M', 'F', 'M'],
        'Age': [25, 30, 22, 27, 29, 28],
        'Salary': [50000, 60000, 40000, 55000, 70000, 45000]}

df = pd.DataFrame(data)

pivot_table = df.pivot_table(index='Gender', values='Salary', aggfunc='mean')

print(pivot_table)

```
